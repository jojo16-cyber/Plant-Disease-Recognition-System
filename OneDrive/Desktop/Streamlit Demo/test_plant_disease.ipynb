```python
import cv2
img_path = 'C:\\Users\\khushbu\\OneDrive\\Desktop\\Plant_Disease\\New Plant Diseases Dataset(Augmented)\\test\\test\\PotatoEarlyBlight1.JPG'
# Reading Image
img = cv2.imread(img_path)
# Convert image from BGR format to RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Displaying Image
plt.imshow(img)
plt.title("Test Image")
plt.xticks([])
plt.yticks([])
plt.show()
```

Neural Network will take only array so converting image into array.

Testing Model

```python
image = tf.keras.preprocessing.image.load_img(img_path , target_size=(128,128))
input_arr = tf.keras.preprocessing.image.img_to_array(image) # Converts image to array
print(input_arr)
print(input_arr.shape)
# Since our model is trained in batches 
input_arr = np.array([input_arr]) # To convert single image to a batch.
print(input_arr)
print(input_arr.shape)
```

```python
# These are probabilities that iamge belongs to a given class or not.
prediction = model.predict(input_arr)
prediction, prediction.shape
```

```python
result_index = np.argmax(prediction)
result_index
```
